```python
# Path to your annotation folder
annotation_folder = '/Users/astha/Desktop Clone/Minor/YOLO Object Detection/data/data_images5'

# Function to extract the class name from the filename
def extract_class_name(filename):
    # Remove file extension
    filename = filename.split(".")[0]

    # Split by "_" and filter out numeric parts and known suffixes
    parts = filename.split("_")
    class_name_parts = []

    for part in parts:
        # Ignore numbers and common suffixes
        if not part.isdigit() and "jpeg" not in part.lower() and "jpg" not in part.lower() and "rf" not in part.lower():
            class_name_parts.append(part)

    class_name = "-".join(class_name_parts)

    # Remove "-png" suffix if present
    if class_name.endswith("-png"):
        class_name = class_name[:-4]

    return class_name

# Dictionary to store unique class names
unique_classes = {}

# Loop through all XML files in the annotation folder
for xml_file in os.listdir(annotation_folder):
    if xml_file.endswith(".xml"):
        xml_path = os.path.join(annotation_folder, xml_file)

        # Parse the XML file
        tree = et.parse(xml_path)
        root = tree.getroot()

        # Extract the filename from the XML
        filename_tag = root.find("filename")
        if filename_tag is not None:
            filename = filename_tag.text
            class_name = extract_class_name(filename)

            # Update all object names
            for obj in root.findall("object"):
                name_tag = obj.find("name")
                if name_tag is not None:
                    name_tag.text = class_name  # Set correct class name
            
            # Save the modified XML file
            tree.write(xml_path)
            unique_classes[class_name] = unique_classes.get(class_name, 0) + 1
            print(f"Updated: {xml_file} → {class_name}")

# Print final class counts
print("\nFinal class distribution:")
for name, count in sorted(unique_classes.items(), key=lambda x: x[1], reverse=True):
    print(f"{name:20} {count}")

print("\nAll annotations updated successfully!")
```

```python
# Read xml file from each file we need to extract
# filename, size(width, height), object(name, xmin, xmax, ymin, ymax)
def extract_text(filename):
    tree = et.parse(filename)
    root = tree.getroot()

    image_name = root.find('filename').text
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text

    objs = root.findall('object')
    parser = []
    for obj in objs:
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = bndbox.find('xmin').text
        xmax = bndbox.find('xmax').text
        ymin = bndbox.find('ymin').text
        ymax = bndbox.find('ymax').text
        parser.append([image_name, width, height, name, xmin, xmax, ymin, ymax])

    return parser   
```

```python
#Assign no to objects(Label Encoding)
def label_encoding(x):
    labels = {
        'Handgun': 0, 'Knife': 1, 'Automatic-Rifle': 2, 'Sword': 3, 'Grenade-Launcher': 4,
        'Sniper': 5, 'SMG': 6, 'Bazooka': 7, 'Shotgun': 8
    }
    return labels[x]
```

```python
train_df['id'] = train_df['name'].apply(label_encoding)
test_df['id'] = test_df['name'].apply(label_encoding)
```

```python
#Save images and labels in text
import os
from shutil import move
```

```python
train_folder = 'data_images5/train'
test_folder = 'data_images5/test'
os.mkdir(train_folder)
os.mkdir(test_folder)
```

```python
cols = ['filename','id','center_x','center_y', 'w', 'h']
groupby_obj_train = train_df[cols].groupby('filename')
groupby_obj_test = test_df[cols].groupby('filename')
```

```python
groupby_obj_train.get_group('Automatic-Rifle_3_jpeg.rf.9c1552f182cf1c1c9117ddbc8daf1f85.jpg') #.set_index('filename').to_csv('sample.txt', index=False, header=False)
```

```python
#Save each image in train/test folder and repective labels in .txt
def save_data(filename, folder_path, group_obj):
    #Move image
    src = os.path.join('data_images5',filename)
    dst = os.path.join(folder_path,filename)
    move(src,dst) #Move image to the destination folder
    
    #Save the labels
    text_filename = os.path.join(folder_path,
                                 os.path.splitext(filename)[0]+'.txt')
    group_obj.get_group(filename).set_index('filename').to_csv(text_filename,sep=' ',index=False,header=False)
```

```python
filename_series = pd.Series(groupby_obj_train.groups.keys())
```

```python
filename_series.apply(save_data,args=(train_folder,groupby_obj_train))
```

```python
filename_series_test = pd.Series(groupby_obj_test.groups.keys())
filename_series_test.apply(save_data,args=(test_folder,groupby_obj_test))
```
